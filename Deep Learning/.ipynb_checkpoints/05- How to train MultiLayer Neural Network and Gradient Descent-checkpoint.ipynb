With forward propogation and doing 2 steps at each hidden layer neuron we get the predicated output y^. As we already know the output from train data we campare these 2 output using a loss function. Our Aim is to reduced the loss hence using optimzer we update the weights in back propgation. One of the known optimaztion technique is gradient descent. Gradient Descent will update all the weights such that y^ will be equal to y. Calculating the derivative means we need to find the slope (slope is a line passing through a point).



Lets suppose we consider our intial weight was the red mark as shown in the picture. The derivative will help find if the slope is +ve or -ve. In this case we will say that the slope is neagtive, IN order to find the slope/derivative we will just draw a line through this point. Then we check if the right hand side on the line is pointing down or up, if its down then -ve slope and wise versa. the aim of gradient descent is to achive the lowest point called global minima. Until we reach global minima updation of the weights will happen through back propogation. If our loss function is not reducing then that means our points is moving around the curve and are not reaching global minina.

If our point is on negative value i,e our derivative is negative then we need to add this derivative to the weight so that we can move towards the global minima so we need to do weight - learning_rate (-dloss/dweight) so -ve * -ve becomes positive. The rate at which we want to move towards the global minina is decided by learning rate.
