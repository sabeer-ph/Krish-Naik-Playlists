IN the example we take 3 hidden layer neural network. We know that the sigmoid values ranges from 0 to 1 and derivative ranges from 0 to 0.25. Howevr this problem can happen with any activation function. When we have weights initiazed as high then while caculation the derivative multiplying these high values will give a even higher value.
