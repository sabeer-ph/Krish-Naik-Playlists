There is one problem with ReLu also. Suppose we are calculating the derivative using the chain rule and we know that the derivative is either 1 or 0. so what if we get one zero then the full derivative value is zero.  so it creates a dead neuron.



WE can solve this problem using Leaky ReLu

In Leaky ReLu instead of saying the value will be between 0 and +ve. We add a small value to 0 we dont keep it as zero. The reason is while taking the derivate of this small value it wont be zero and we can fix the problem with dead neuron.
