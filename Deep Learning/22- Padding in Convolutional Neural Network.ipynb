

**Padding :**  we multiplied a 6 x 6 image with filter of 2 x 2 giving out a final image of 4 x 4. but there is some amount of information we loose because of the reduction is size, If we want the output size to be same as input then we use padding.

We know the formual for calculating the output size from input. Now that we know what are output size should be we caclulate what should be the size of the input and add padding to make it that size.

n - f + 1 = 4 (intially)

n - f + 1 = 6 (now to want this)

n - 3 + 1 = 6 --> n = 6 + 3 - 1 = 8 (we want the input image to be of size 8)

to make the input 8 x 8 we make padding of 1 on each side

so what values do we fill in these padded places. 

    1. 0 padding : fill zeroes (most commonly used)
    2. Neighbour value : fill the same value as neighbour
    



IN our case we do zero padding and the new formual is calculate is to just add 2p where p= 1 padding value.
