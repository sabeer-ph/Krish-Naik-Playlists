The simple concept of Data Augmenation is that if we one image of a cat then using data augmentation we can transform these images to different types of images i.e by applying transformations like flip (invert), horizontal shifting, vertical shifting, zooming and zoom out. These images are then given to CNN to predict the output. 


**Advantage :** when we dont have enough  data then with Data augmentation we get more data (transformed data) 

The CNN model we create shoudle be **ROBUST** that means for example if the cat in our image is upside down still we should be able to identify it.
