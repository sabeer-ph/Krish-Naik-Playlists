

**Convolution Operation :** We have an input of 4 x 4 and filter of 2 x 2. If we apply the filter on the input n - f + 1 --> 
4-2+1=3  gives 3 x 3 matrix. this full process is our convolution. 

Once the Convolution Operation is completed we can stacked multiple Convolution Operation or use **Max Pooling**

Suppose Our image includes 3 cat faces and our filter is used to find the faces. As we go more in high level these faces should be more accurately and precisely detected for this max pooling is used.

**How Does it work?** Max pooling is like another filter matrix which will be applied on to the output image from convolution. Then the max value in the overlapped place is taken and made an output **(remeber the stride = 2 i.e jump from overlapping)**. So if the convolution output image slightly identifies the shape of face then the max pooling will give more clear picture as we are taking max. 

There are different types of pooling like **Mix** and **Avg**.
