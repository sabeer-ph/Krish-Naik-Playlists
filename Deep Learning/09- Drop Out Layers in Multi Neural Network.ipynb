In random forest we create multiple decision trees,each and every decision tree will be  creating a decision tree to complete depth when we say complete depth means leading to an over fitting problem. 

Random forest has  one more technique where we will not be using all the features in random forest. i.e just using a sample or a  subset of features we will be using and is our  regularization method which will  actually reduce the overfitting problem. and improve the accuracy.

random forests is discussed here because we create decision tree with the help of  subset of features and we create  multiple decision every time creating a new decision tree at that end we will have some different subset of features. Similarly we implement the dropout.





suppose this is our sample neural network, first layer is input layer the second layer is hidden layer and finally output layer. To implement the dropout first we select **dropout ratio which is in between 0 and 1**. Like in random  forest selected sample or subset  of features similarly in NN selecting subset in input layer, subsets of activation function in hidden neurons and selecting subset of neurons in output layer. we will not  be selecting everything we will just be selecting subset of features in each and every hidden layer along with the input features.

So in our example you can see that we have selected two input features and the p value for the  first layer 0.5, this p value is nothing but dropout ratio. Similarly in hidden layer 1 two neurons are activated remaining all are deactivated so again p value almost equal = 0.5 (coz 2 out of 5 are activated), similarly hidden layer 2 two are inactivated and two are  activated so here also p = 0.5. 

So we have created a neural network and selected dropout ratio of P = 0.5 in each and every layer. When forward  and the backward propagation will be  going with p = 0.5 some features in input layer will deactivate. suppose in this case second and the fourth mode is  deactivated, In first hidden layer selected  p-value is 0.5 so randomly it will  select some of the activation function  some of the neurons will  deactivate finally we'll get the  output with respect to the first  propagation and the backward  propagation is same the activated neurons weights will  get updated.

Now in the  next titration  p-value = 0.5 half of this node in the  input feature will again be activated, it'll be selected  randomly every time with respect to this  p value.
