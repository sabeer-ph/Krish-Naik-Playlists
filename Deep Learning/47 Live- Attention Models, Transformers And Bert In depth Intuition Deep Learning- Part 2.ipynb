**Attention Based Model :** We need a model where it knows the relationship with the last word and the next word. So we use a bidirectional RNN in the Encoder. we given the same data to the forward direction and reverse direction (My name is sudhanshu). So in output is the combined output from each cell [boxes with contains NN].



We can try to sent the combined output from the encoder to the decorder but then it becomes one to one mapping but we dont need that. 

so what we do is we use attention based model in that we use NN in between encoder and decoder and with this we try to get some key words with which we can get the context of the sentence.

For example lets take an example setence "The Bull is running high". So if we just give inportance to bull and running the it means the animal running but when we give importance to bull, running and high then it means increase in stocks.



SO instead of just using Encoder and Decoder we use a NN inbetween them which will focus (give attention) to the words and try to find the key words which keys meaning to the whole senetence and this is called attention based model.

The number of input we can use to NN is an hyperparameter which we can calculate. In the Reserch paper they have tried with diff number of inputs and also provides the respective accuracy.
