We will try to understand whats Convolution in CNN. 

first lets understand basics of Images. An image can be of Graye Scale or color and it units in **Pixels**. If the Image is of graye scale (means just black and white) then for example we say 6 x 6 image.

If the image is of color then its made of 3 color RGB then we say its 6 x 6 x 3 image. in RGB also each color value ranges from **0 to 255**

Lets consider an example



On the left hand side we have a 6 x 6 graye scale image in which **1 is white and 0 is black**.

As dicussed in the previous section in the visual cortex of human brain we have different layers which does some task and in CNN its called filters. So filter does task like finding the egdes. 

Here in this example we use 3 x 3 filter which is **vertical edge filter** which finds the vertical edges in the picture on the left. the vertical edge in our picture is between 3 and 4 columns where the zeros change to one. So when multiply an image of 6 x 6 with a filter of 3 x 3 we get a 4 x 4 image. lets see how



**step 1 :** place this 3 x 3 over the 6 x 6 starting from the top left then multiply each term of the filter with input image and add them up. i.e 1 *2 + 0 * 0 + -1 * 0 + 2 * 0 + 0 * 0 + -2 * 0 + 1 * 0 + 0 * 0 + -1 * 0 = 0 . thats the first value in our 4 x 4 matrix.

**step 2:** move one step to the right of the input image and place the filter and redo the multiping and summing we get -4.

**step 3 :** once we complete the first line then we move one step down and start from left.



Next we scale using minmaxscaler the output i.e is -4 will become zero (white) and 0 will become 255 (black). hence we can find the vertical edge.



This multiplication of input to filter is called **convolution**.

lets says our input image size is **n = 6**

filter size **f = 3**

then the output size is calculated by the formula **n - f + 1** in our case **6 - 3 + 1 = 4**
