Why use Recurrent Neural Network (RNN) ?.

RNN works well when the input is sequence of data. Sequence of data means for eg we have a sentence and we need to predict if positive or negative. 

Some applications : Spam classifier, time series data, sales and stock forecasting can be achieved with a good accuracy using RNN. 

in NLP we learnt BOW, TD-IDF and Word2Vec and in all these cases input text is converted to vectors. In all these 3 cases sequence information is discarded. Once the sequence is discared then the meaning is lost then the accuracy will be on the lower side. 

When we ask questiion to google assistant and alexa then the answer is meaningful in a sequence of text. If we want to control the sequence of text then we need to use RNN. 



Another application of RNN where it is extensively used is time series data. Lets consider an example of sales of a data in time series. And our task is to predict a particular time what is the output. In Machine learning the usual financial techniques are
**ARIMA & SARIMAX**. This these 2 techniques we will use the time window of previous days and predict the value. RNN is veryful to predict the output in time series data.

Some of applications where RNN is already is **Google Image Search** 

**Image Captioning** where we provide an image then the app gives us suitable caption for it.

RNN in used in **Google Translator** where when we start to type it starts to translate it to another language (Technique used is many to many RNN).
