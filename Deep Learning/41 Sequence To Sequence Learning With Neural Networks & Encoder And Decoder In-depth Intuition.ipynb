In this section we will try to understand Sequence to Sequence with NN, Encoders and Decoders with its working.

Sequence to Sequence means where our input is sequence info and also our output is sequence.

Eg 1 Language translator : We will input data as sequence in one language and output is another language in sequence.

Eg 2 Image Captioning : We give input as image and output is a text as caption for image

Sequence to Sequence  has 2 main components

    * Encoders
    
    * Decoders

This example is of Language Translator to convert English to french we give input A,B and C in English and we Get X, Y and Z in French



In sequence to Sequence the NN can be RNN, LSTM or GRU but mostly LSTM is used as it works well. The ENcoder part of LSTM NN outputs we will removed except for the last timestamp (remeber that our x axis is time). the final one is at tn AKA end of stream.

The the output of a Encoder is vector AKA **Context Vector**. the input we provide is also vector we use methods like one_hot, embedding and word2vec to convert the input to vector. 

This context vector is given as input to Decoder and we get the output X which is out train data then this is given as input to get output in French.

the output y^ is calculate using probabilty compared to the input. We caculate our loss function the use optimizer and update the weights during back propogation.

If our example is of Image Captioning then our encoder will be some advanced CNN (Vgg16, ResNet50) and the output of the ENcoder will be a flattened vector which will be given an input to Decoders.

**Disadvantage :** When the length of the sentence increases then the accuracy descreses. The translation will not happen properly. We can try to overcome by using Bidirectional LSTM in Encoders.
