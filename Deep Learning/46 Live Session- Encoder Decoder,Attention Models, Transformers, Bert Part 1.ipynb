**Problem Statement :** There were many problems people were planning out of which 2 main ones are.

    * Translating long sentences from one language to another.
    
    * Underdtanding the context of the sentence in chat bots ( bull can mean animal and also upward trend in finance.

people have done lots of research in CNN. In CNN many other archirecture were also release like VGG16,VGG19, RESNET50, RESNET34, RESNET101, INCEPTIONNET, GOOGLENET, DARKNET . Bu these CNN based NN will not be able to solve the problem statements above.

**Why ?** in CNN we give input from one side and we get output. we calculate loss function and using optimiser during back propogation we update the weights and this continues for each iteration. The **problem** here is in the 2nd iteration CNN doesnt remember what input was sent in 1st iteration.

CNN are good to interpret image and videos i.e object detection, object tracking, object segmentation. 

But when it comes to text as input we need to remeber the last words and understand the context of the sentence.


SO we wanted a network that not just understand the **weights** and **relationship between data** but also understand the context and could remember something from past and able to changed the context based on the input we give.this derived network is called **LSTM**.

In LSTM we have RNN and GRU(Gated Recurrent NN). so in here we have input and output and in addition we have input a feedback from output to input.

People were still facing problems these NN **alone** were not good enough to solve the problem statemnet we have. The model has to understand the complete data/ setence and then translate it on in case QA system provide an answer. 

so a reserch paper was released called [Sequence to Sequence](https://arxiv.org/pdf/1409.3215.pdf) which could solve the problem to certain extend.

so there was idea to see what will happens if we separte the input and output will the accuracy be more in terms of langauge translaltion and QA program. Based on this idea Sequence to Sequence paper was released. 

So the input and output were separted and input was called **Encoder** and output was called **Decoder**.

    * Encoder : Is an input network which takes input and understands the relationship between i/o based on the loss ( y - y^).
    


How to train this network. SO we will have many english sentences (Xi) and we will have out respective matching Hindi sentences (Yi). We cannot give these senetnces directly to the model. We need to have numerical represntation. So we do Embedding ( other methods like Word2Vec, Tf-Idf can also be used).

With the output recived we can calculate loss function and then we can back propogate and update.

**Problem :** In order to understand the context of a sentence the word before and word next plays a very inportant role. But in the case in our encoder the data propogates in one direction. This encoder network will not be able to understand the data in both the direction. 

**Solution :** is to use a bidirectional LSTM in encoder
