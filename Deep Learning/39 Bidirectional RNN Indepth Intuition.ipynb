Bidirectional RNN is next step towards advancement. 

Lets consider a simple RNN. RNN is unidirectional



Here we have inputs x11, x12,x13,x14 at time t, t+1,t+2,t+3 and we also have ouput y11^, y12^, y12^ and y14^. so inorder to calculate the output y13^ we need to know the inputs x11, x12,x13. But what if the output of y13^ is dependent on the future. for eg : He likes to eat ___ ? . we want to find the next word is it apple, banana. So in RNN we cannot get the info from future in order to solve we have a concept called **Bidirectional RNN**.

**Bidirectional RNN :** We will take the same RNN we called it backward layer RNN and we reverse the direction of information. in other words we are adding an extra LSTM layer and passing the data in the reverse direction.



from the above picture lets understand we have taken the same RNN (in Blue Color) and we reverse the flow of information.  Now the last one and input x14 and it passes to next layer with input x13. the output from these are joined to the output y11^, y12^, y12^ and y14^ and since it a clasification problem we use a sigmoid there marked as circle.

Now if we want to caculate y13^ we need to know the inputs x11, x12,x13 and we also know x14 from the reverse RNN.

**Bidirectional RNN Uses :** can be used with most of NLP uses cases expect for speech recognisition.

**Disadvantage :** As we have 2 directions of information flow its very slow comared to simple RNN/LSTM RNN.
