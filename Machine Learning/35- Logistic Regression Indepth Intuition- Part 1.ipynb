In This section we try to understand why linear regression cannot be used for binary classification problem.

So we have a training dataset of "Not Obese" (0) and Obese (1) people. We plot these value on a graph and using linear regreession we fit a best line with less error. So 75 is our threshold value and draw a line it will meet the best fit line at 0.5. So we consider anything above 0.5 is Obese and less than is Not Obese.

**Problem :** If we add one outlier to the Obese dataset then the best fit moves and it become more slant. So now the threshold is moved to more than 80 which is not we wanted. Another problem if we take small and very high weights and if we draw a vertical line to the best fit line then we either get a value higher than 1 or values which is -ve. 

Hence we use Logistoc Regression which we will see in next part
