So we have a dataset of height and weight, We create a histogram of the weight distribution. So in Probability Density Function (PDF) we smooten the bar graph of histogram to a curve like structure. In this case it will be bell curve. 

Note: in histogram the Y axis is **count**.

So in PDF we smooten the curve and change the Y axis to **% of Distribution**



So we take any point on the curve then it the percentage of the distrbution (from left to that point) from the total distribution). Always the total distibution will be 100 (i.e 1). If we take a point at the exact centre then its 50% or 0.5.

### Cumulative distribution function (CDF)

We can build a CDF using a PDF. As the name says its cumulative(adding up each value with all previous ones). 

So we have PDF, in order to build a CDF we add each point with all its previous values and we start plotting the new graph. This graph starts from the left point and starts going up to the highest point which is 1 (100%). Now lets take any point on the CDF and find its X value (130) and then find its Y values (0.9) so this mean 90% of our weight values are less than 130. which also means 10% of our values are greater than 130.
