In this section we will try to understand Underfitting, Overfitting, Bias and Variance ( Also high and low Bias & variance)

Here we take example of Regression with degree of polynomial =1, 2 and 4.



    1. degree of Polynomial = 1 is simple regression where best fit line is a straight line.
    2.degree of Polynomial = 2 is not a perfect line with bit of curve.
    3. degree of Polynomial = 4 perfectly fits the curve.

Polynomial = 1 the error between points and best fit line is high hence its underfitting. Hence the model doesnt work for the training data and also for the test data. This model we call it as high Bias (error WRT training data is high) and High Varaince (error WRT to testing data is high) model 

Polynomial = 4 the lines pases through all the points and hence zero error. This is overfitting. Here the model works best for the training data but not for the test data. This model we call it as low Bias (error WRT training data is low) and High Varaince (error WRT to testing data is high) model. 

Polynomial = 2 is better than the above both cases where we have low bias and low varaince.

**Decision Trees :** Even here we have problem of overfitting where we split our data till the end. To reduce this we have methods like pruning and max depth. Also the other solutions is **random forrest**.

In Random Forrest we create multiple Decision tress in parallel. We create these using random features not selecting all of them at once. Then we aggreagte the data from them i.e if most decision trees say output is 1 then we select 1 else 0. Even though decision tree has low bias and high varaince when we taken them in parallel and aggregate the high varaiance will become low.
