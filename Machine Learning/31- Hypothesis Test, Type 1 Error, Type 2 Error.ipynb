Statistics is all about data and there's a huge amount of data and the data will only be useful if we work to analyze it, find out some conclusions from it.

So it is very much important to find out such an important  interpretation and conclusion and for that we basically use hypothesis testing

**Hypothesis Test :** We evaluate 2 mutually exclusive statement on **population data** using **sample Dataset**

Our intitial Hypothesis is called Null Hypothesis (H0) and the exclusive hypothesis is called Aternate Hypothesis (H1)

We collect data and analyse the data and based on data we either accept or reject the hypothesis.

**Type 1 Error:** We reject the null hypothesis with lack of evidence

**Type 2 Error:** We dont not reject the Alternate Hypothesis.

We reject the null hypothesis when the P < 0.05 where P is the significance Value.
