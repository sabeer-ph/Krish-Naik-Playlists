This is first step in Explorartory Data Analysis(EDA) is to do Univariate, Bivariate and Multivariate Analysis. This helps us to analyse more about data, analyse data and leads us in the right path to implement machine learning.

first of all in all 3 cases we use continous data (numercial) data for plotting, we can classify the plotted continous data using the categorical features.

Lets consider a dataset of height and weight and with its output of 3 categories.



**Univariate :** here we take one continous feature and plot on the x axis and then we try to classify/cluster the data



**Bivariate :** Here we take 2 continous feature and plot against each other. 



When we have 3 features we can built a 3d plot but its will be hard to analyse it, Also more than 3 features as humans it will be hard to plot and analyse it. so we do multivariate analysis.

**Multivariate :** Here what we do it we plot one feature with every other feature we have so at a time we only take 2 features. 



In order to built a multivariate plot we have have a function in seaborn i.e sns.pairplot().

using bivariate and multivriate by comparing 2 features we could find the **Corelation** between them. 

**Corelation :** When comparing 2 features in a scatter plot, is one feature increasing or decresing with respect to the other. In this we have **3** types.

    * +ve : When one feature increases the other increases too.
    * -+ve : When one feature increases the other one descreses.
    * zero : when one increases or descreses the other one no change.
    
the value of corelation is between -1 --> 0 --> 1
