**Histogram :** this tells us how many points (count) is there within a range or also called bin size. Histograms are built on single feature.

In univariate analysis we have our plots on x axis and our y axis is zero. However its hardto find out how many people weights are within each bin size. thats were histogram comes up. 



We can built a histogram using plt.hist() function. 

if our histogram looks like a bell curve (gaussian distibution, normal distribition) we can convert it to probability desnity function PDF. Once we do that we know what percentage of points are there in each range. we will expore it more in next session.
