```python
import statsmodels.api as sm
df_adv = pd.read_csv('./Exercise_Data/28_2 Advertising.csv', index_col=0)
X = df_adv[['TV', 'radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
# formula for mullti regression problem is y = c + m1x1+ m2x2+ m3x3 as we have 3 features tv, radio and newspaper. 

X = sm.add_constant(X) # we created a column with value 1 for intercept "C"
```

```python
X.head()
```

 ordinary least squares (OLS) is a type of linear least squares method for estimating the unknown parameters in a linear regression model. 

```python
## fit a OLS model with intercept on TV and Radio

model= sm.OLS(y, X).fit()
```

```python
model.summary()
```

coeff for Newspaper shows -ve value. Which means newspaper add is not affecting the sales. Also its P value is greater 0.05. we can drop this feature.

R-squared value will be between 0 to 1. More its near to 1 means our model has fit well. in our case above its 0.897.

```python
import matplotlib.pyplot as plt
X.iloc[:,1:].corr()
```

The above correlation between features shows no much corealtion that means we need each feature to predict our output

```python
df_salary = pd.read_csv('./Exercise_Data/28_2 Salary_Data.csv')
df_salary.head()
```

```python
X = df_salary[['YearsExperience', 'Age']]
y = df_salary['Salary']
```

```python
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
model= sm.OLS(y, X).fit()
```

```python
model.summary()
```

```python
X.iloc[:,1:].corr()
```

The above 2 features are 98% corelated. So we can say that one of the feature is enough to predict the output. WHich feature to removed as be decided at P values. The P value for Age as shown above is 0.165 which is way greater than 0.05 so we can drop this feature.
