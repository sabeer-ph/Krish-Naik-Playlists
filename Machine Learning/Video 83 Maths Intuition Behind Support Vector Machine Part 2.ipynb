Here we will try to understand the maths intuition behind support vector machine. and the main focus is basically to get the maximum margin distance so that we'll be able to separate the points quickly.

to begin with lets understand basic difference between SVM and logistic regression.

In LR we only draw a hyperplane (3D) or a line (2D) to separate the points but in SVM we have 2 extra hyperplane/line to separte the data into 2 classes.



first lets understand how does the points above and below the lines becomes +ve and -ve



for example lets take a line that passes through "0". the slope (W^T) is -1 and intercept is zero

our equation for the line is y = W^T + b where b = 0. hence its y = W^T 

lets take many points below the lines and one of the point is (-4,0). so if we try to compute y for this using matrix multiplation it becomes as shown in the above figure 4. So all points below the lines are postive (lets consider it as +1).

next we will take many points above the lines for eg one point (4,4) and if we compute the y we get -ve value. so all points above the line is -ve.

next from this line we draw a line to the nearest point to both the classes.once we get the nearest point we draw a parallel line.

SO we have 3 equation now for the -ve points its w^T+b = -1 and for the +ve points its w^T+b = +1. So to get the distance between these 2 lines we subtract one from another

so we w^T (x2 - x1) = 2. inorder to remove W^t we divide by normalized ||W||. so our aim is to increase 2/||w||. that is get choose the one with the maximum distance.

SO the last formula in the above picture we understand yi = +1 when w^Tx + b >= 1 and yi = -1 when w^Tx + b <= -1 so we group these formula as one. 



+1 * 1 = 1 also -1 * -1 = 1 so that is how we derived the above formula. So if we get any value less than +1 that means that point is mis classified.

in reality we cannot perfectly classify all the points , there will be some amount of mis classification. Hence the extra part added in the above formula.

**Note :** all these works when the data is linearly separable i.e using straight line.
