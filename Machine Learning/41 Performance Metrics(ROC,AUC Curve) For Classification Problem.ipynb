ROC AUC is a performance metrics used for Binary Classification problem.

By default in Logistics Regression it considers anything above 0.5 as 1 and below it at 0 but this thresold value can differ in other use cases. 

for example in heath care uses cases like disease prediction this thresold value will play a very important role. Bases on less True positive rate / False positive rate we call change the threshold value.

We can show how well a model is performing using a graph to the domain expert who will help us to decide what thresold value is the best.

So as shown in the above picture we take different threshold values [0,0.2,0.4,0.6,0.8,1] and we caculate the new ouput bases of the y probability. if probablity is > thresold value then 1 else 0. if do this with each and every threshold values.

Once done then we calcuate the TPR and FPR and with values we plot the **ROC curve**. The **area under the ROC curve is AUC**.
