While doing a train test split we choose a parameter called random_state and every time we change this value the shuffling of data happends and the final accuracy changes. So what do we do?

In order to overcome this problem we have Cross Validation technique. Lets understand different types

Here every time the test is done with one dataset and rest is used for training and this is repeated with each dataset. So if we have 1000 records then everytime we have 999 train and 1 test dataset.

**disadvanatge :** lot of compulation needs to done i.e if we have 1000 records then we need to run this 1000 times.Second : this is low bias i.e works well with training dataset but not with new test dataset.

Note: This is an old method and not used in current times.

**2. K Fold CV**



so if we select K = 5 then we will run the experiment for 5 times and each time we take the test dataset = 1000/5 = 200. So we 5 accuracy score and we calcuate the mean of it as our accuracy. The range of this final accuracy can vary from average(min(5 values)) --> average(max(5 values)).

**disadvantage :** SO when we take 200 as our test dataset what if this 200 records has mostly one type of data (i.e if its a classification problem and most out outputs are 1. SO it will not give us exact results.

**3. Stratified CV**



This model is same as K fold but this overcomes the problem K fold has. So when the dataset is split for train and test it makes sure that equal amount of data from each instance/class will be used in the train and test dataset.
