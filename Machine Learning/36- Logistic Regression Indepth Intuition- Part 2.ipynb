Logistic Regression is used for binary classification and we can also modify it for multi class classification. However, in this section we will focus on binary classification.

Logistic Regression can be applied on a problem statement where 2 class can be linearly separable (means it can seperated with a straight line). 

As shown in the plot we plot the blue and red values . in LOgistics regression we consider blue values as yi=1 and red values yi=-1. We know that equation of staright line is y = mx+c (also y=wTx + b ). Single our straight line passes through zero our intercept is zero hence the formual becomes y = mx (where m is the coefficient we need to find)

our straight is a postive slope(the right side of the line is poiting upwards so any points above the like is postive. SO we used this idea to know if our points are correctly clasified.

Case 1 : blue point above the line : yi * WTxi --> +ve * WT * +ve > 0. So our point is correctly classified

Case 2: red point below the line. distance from line to point is -ve and the point value is -ve --> -ve * -ve gives +ve > 0. Hence correctly classified.

Case 3: red point above the line. distance from the line to point is +ve and the point is -ve --> +ve * -ve < 0. incorrectly classified.

Case 4: Blue point below the line. distance from the line to point is -ve and the point is +ve --> -ve * +ve <0 incorrectly classified.

So we have a cost function which the summation of all the points which its distance. and we need to find the co effieinet (shown top right of the picture). So there will be so many lines we can try but we choose the line as best line where the **cost function is maximum**.

**important :** but this cost function will not work well when we have a outlier. When we have a outlier then the best fit line will shift so in order to overcome that we modify our cost and add sigmoid to it. So no matter how far our outlier is the value of it will be changed between -1 to +1.
