**Hypothesis Test :** Evaluates 2 mutually exclusive statement on Population dataset using sample dataset

**One Sample Proportion Test :** 1 categorical feature.

Example : Lets take one categorical feature **Gender (M & F)** and we want to find out the propotion of Male and Female in population dataset using the sample dataset we have.

Ho : There is no difference in propotion
H1: there is difference 



With the sample dataset on gender we have if we plot a bar graph we can know if the dataset is equal or not but thats not enough coz the data we have is just sample and not population. So in order to confirm we do a test which is One Sample Propotion Test. If we get a P value <0.05 (2.5% at the 2 tail ends of normal distribution) then we reject null and accept alternate hypotesis.

**CHI Square Test :** 2 Categorical Feature Gender and Age group and the Question is a difference in Gender based on Age group.



**T Test :** One continous Feature. Lets Take weight, The question is there a difference in Mean Weight of current sample with the previous sample. 



Note : T Test is also done when we have one categorical and one continous feature but the condition is that the categorical feature should not have more than 2 data (for eg just Male and Female). What if we have more than 2 then we do ANOVA

**Corelation Test :** We take 2 continous feature here height and weight and try to find the realtion between them. FOr this we use Corelation test where the value range between **-1 --> 0 --> +1**. -ve --> no relation --> +ve 

**ANOVA TEST :** with one or more categorical data and 1 continous feature we do ANOVA TEST but the conditional is the categoorical data should more than 2 values (else T Test)
