**independent Event :** 2 events are independent if their probability doesnt affect each other.

**Eg :** Tossing 2 coins where their probabilty of getting an head or tail is not affected.

**dependent Event :** Event 1 probabilty affects event 2 probability.

**Eg :** we have a bag with 5 marbles out of which 2 are black and 3 are Red.

Event 1 : We pick a marble then whats the probabilty of getting a black = 2/5

Event 2 : out of the 4 left whats the whats the probabilty of getting a black = 1/4



**Conditional Probability :** whats the probabilty of taking out 2 black marbels back to back.

Event A : P(A) --> taking one black marble = 2/5

Event B after Event A has already taken place P(B | A) = 1/4.

the Total probabilty of both event A & B is P(A  B) = 2/5 * 1/4 = 1/10

the formula for conditional probabilty is  p(b|a)= p(anb)/p(a)
