The equation for the best fit line is y = mx + c . where x is the slope and c is the intercept.

**Slope :** when x = 0 then y = c. that is at x=0 at what point on c does the line pass through.

**intercept :** Whats the change in value between 2 points on the slope. for Eg lets take 2 point on X 1000 and 1100 to want to know whats the change of value on the line.

**How to find Best Fit line :** we start with a line then draw lines from the points to the line, then sum all these lines. then we move the line in any direction then calcuate again the sum of all the lines from points to the line. if its more than the previous calcuated one then we moved in the wrong direction. We keep doing this until we get the lowest value of sum of lines from points to the line.

**m :** No of the points in the graph.

**y^ :** is the point on the best fit line.

**y :** is each point in the graph.
