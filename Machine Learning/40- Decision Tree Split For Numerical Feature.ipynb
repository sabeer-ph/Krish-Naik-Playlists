In this section we will try to understand how does the split happens in decision tree if the input feature is a numerical value. 

Step 1: Sort all the input feature.

step 2: consider a threshold value. The first value 2.3 is taken as thresold and a split is done. As we see in the picture we get 1 YES/ 0 No on the left and 3 YES/ 4 No on the right.

STEP 3 : repeat step 2 with next number 3.6. in all cases Entropy + information gain is calcuated and the one with the best is chosen.

Disadvanatge : When we have million of records the this task of going through each number and spliting will take a lot of time.
