**Entropy :** Is the measure of purity of the split and its value ranges between 0 to 1 where near to 0 is good split and close to 1 is bad/impure split.

Here in this case we have a classifcation problem wirh 9 YES and 5 No and we have 3 feature f1, f2, f3. So Our aim it to reach the leaf nodes as fast as possible for that we need to choose the best feature which gives us pure split. Pure split means we just only one class at that point (YES Or NO).

So lets calculate the Entropy for the point f2 with 3 YES and 2 NO with the given formuala. 

With Entropy we check the purity at each node but thats not enough, so to overcome that we use **Information Gain** which takes into consideration from the root to the leaf at once.
