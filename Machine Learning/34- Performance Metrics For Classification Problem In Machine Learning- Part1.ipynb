As mentioned Earlier that Accuracy metric works only with balanced dataset and not imbalanced dataset. Below an example on why?

**Imbalanced Dataset :** For example we have 1000 records training dataset out of with 900 belongs to 'Class A' and 100 becomes 'Class B'. So if the model predicts all 1000 records as class A and if we calculate accuary for that it will 900/900+100=90%. So this metric will be wrong.

In case of **Imbalanced Dataset** we use metrics like Precision, Recall and F1 Score.

**Recall --> True Positive Rate(TPR) --> Sensitivity :** From the total actual postives how much did the model correctly predict as postive. Here we should try to reduce the False Negative. **Use Case :** Cancer patient . if a patient was actually had cancer and the model detected as negative (FN) then he will be send home. So in such case we need to use High Recall and reduce FN. 


**Precision --> +Ve Predicted Value :** From the Total postive predicted by the model how much very actually postive. Here we should try yo reduced the False Positive. **Use Case :** Spam email . The model predicted the email as spam but it was ham then it will be sent to spam folder in this case we use precison and try to redcude FP.

**FBeta Score :** A case when both Accuracy and Precison are important for us then we use Fbeta Score.

When and what value to choose for Beta

    1. when both FP and FN are equally important then we choose Beta = 1. This makes our formula as Hormonic Meas (2xy)/x+y.
