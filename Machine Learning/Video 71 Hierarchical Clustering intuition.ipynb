Hierachial Clustering is an UnSupervised Learning used to create clusters based on the distance between the points. The distance between the points is calculated using Eucledian Distance.

Step 1: All individual points are considered as clusters.

Step 2: nearest 2 points are group to one cluster. So we have P1, P2 and P3,P4 and P5,P6 which are grouped together. These groups are plotted on a **Dendogram** bases on the distance. 

Step 3: nearest 2 group are grouped again based on the their distance and marked on the Dendogram.

Step 4: Finally all groups are grouped to become one group.

But the idea it to find no of clusters and split the data based on their similarities but in this case we ended up with just one. So we use **simple Hack** to find ideal number of clusters.

**Hack :** split the groups where the vertical line is not crossed by a horizontal line. So from the dendogram we are able to find that only the line between group (P1, P2) and (P3,P4,P5,P6) is not crossed by a horizontal line and hence we break it giving us 2 clusters.
