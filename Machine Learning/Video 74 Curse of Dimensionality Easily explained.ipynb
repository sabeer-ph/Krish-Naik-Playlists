Dimensionality means Dimension AKA features, Attributes.

So lets say we have 7 models and use different number of features for each model. As shown in the pictures we use 2,5,10,100,200,1000,10000 features each for the 7 models.

So we will noticed that the accuracy score increases when the number of features increases but this will happen only till a threshold number of features after which the accuracy starts decreasing.
