In this section we will try to understand how we can use Logistic Regression to do multi class classification. For that we try to understand a concept of **One Vs Rest (One Vs All)**.

SO lets understand how can be perform multi class.

**Training Data :** We have 3 inputs here I1, I2 and I3 and the alogorithm creates 3 model where its group 2 inputs as 1 in each model.

Model 1 : I1 and (I2+I3) were we consider I1 as +ve and the other 2 as -ve.

Model 2 : I2 and (I1+I3) were we consider I2 as +ve and the other 2 as -ve.

Model 3 : I3 and (I2+I1) were we consider I3 as +ve and the other 2 as -ve.

**Test Data :** we the send the test data to all 3 model and we get back a probabilty from each of them, We choose the model which gives us the highest probabilty. 

Note : The total of all 3 probabilty is 1.

### Exercise

```python
# Load libraries
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
# Load data
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Standarize features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# Create one-vs-rest logistic regression object
clf = LogisticRegression(random_state=0, multi_class='ovr')
```

```python
#Train One-Vs-Rest Logistic Regression
# Train model
model = clf.fit(X_std, y)
```

```python
# Create new observation
new_observation = [[.5, .5, .5, .5]]
```

```python
# Predict class
model.predict(new_observation)
```

```python
# View predicted probabilities
model.predict_proba(new_observation)
```
