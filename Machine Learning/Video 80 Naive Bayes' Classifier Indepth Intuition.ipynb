Naive Bayes Classifier is used for Classification problems and the base of this classifier is Bayes Theorm.

So lets try to understand how can we apply this to classification problem.

So to solve a classification we will have many featutes (x1,x2,x3.....xn) which are our independent feature and we will have a depenedent feature (y).

So using bayes theorm to our classification problem it beomes whats the probabilty of y given the feature (x1,x2,x3.....xn)

P( y | x1,x2,x3....xn)
