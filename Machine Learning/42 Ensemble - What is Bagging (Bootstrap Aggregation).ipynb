**Ensemble :** means combining multiple models.

Lets try to understand bagging, bagging is also called bootstrap aggregation.

In bagging we have a training dataset **D** and lets consider we have 4 models **M1,M2,M3 and M4**. So give a sample training dataset to each of the model and train them (This is called Bootstrap). Then when we have test data we give one row to each of the models and on the output we do a voting clasifier, the major vote will be our output (this is called aggregation). so it total **bootstrap aggreagtion**
