#### Back Propogation Using Pytorch

y=x^2

```python
import torch
```

```python
x=torch.tensor(4.0,requires_grad=True) # intiatlize gradient mean keep track of the changing value of x during back progation
```

```python
x
```

```python
y=x**2
y
```

```python
#### Back propogation y=2*x and x is 4.0
y.backward()
```

```python
print(x.grad) # 2 * x = 2 * 4 = 8
```

```python
lst=[[2.,3.,1.],[4.,5.,3.],[7.,6.,4.]]
torch_input=torch.tensor(lst,requires_grad=True)
```

```python
torch_input
```

```python
### y=x**3+x**2

y=torch_input**3+torch_input**2
```

```python
y
```

```python
z=y.sum()
```

```python
z
```

```python
# backpropagation is finding the deraivative of y with respect to x. 
### y=3x**2+2x---> where x is in variable "torch_input"

z.backward()

# the first row of x is [2., 3., 1.]

# y = 3*2**2+2*2=12+4 (for first entry 2)
# y = 3*3**2 + 2*3 --> 27 + 6 -->33  (2 entry 3)
# y = 3*1**2 + 2*1 -->> 3 + 2 --> 5 (3rd entr)
```
